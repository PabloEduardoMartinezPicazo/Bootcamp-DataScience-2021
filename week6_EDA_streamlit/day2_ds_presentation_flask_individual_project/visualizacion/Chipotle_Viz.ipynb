### Step 5. Create a scatterplot with the number of items orderered per order price

Make sure you get the same labels and title
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
chipo["item_price"]=chipo["item_price"].apply(lambda x:x.lstrip("$"))

```

```python
chipo["item_price"] = chipo["item_price"].astype(float)
chipo
```

```python
sns.scatterplot(data=chipo.groupby("order_id").sum("item_price"), x="item_price", y="quantity",color="green")
```
