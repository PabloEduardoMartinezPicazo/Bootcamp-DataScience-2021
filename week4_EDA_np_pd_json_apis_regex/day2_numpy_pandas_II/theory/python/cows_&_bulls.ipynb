
- [Cows_and_Bulls]

Say the number generated by the computer is 1038. An example interaction could look like this:

  Welcome to the Cows and Bulls Game! 
  
  >Enter a number:

  >> 1234

  >>> 2 cows, 0 bulls

  >> 1253

  >>> 1 cow, 1 bull
  
  ...
Until the user guesses the number or write "STOP"

```python

num = "1234"

cow = 0


while cow < 4:
    numero=input("numero")
    cow = 0
    bull = 0
    contador=0
    for i,value in enumerate(numero):
        if value in num and value == num[i] :
            cow+=1  
        elif value in num[contador]:
            bull+=1
            contador+=1
        if numero == "STOP":
            print("STOP")
            break

    print(f" you got {cow} cows and {bull} bulls")


```
